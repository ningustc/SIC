```python
import math
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import signal
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import AxesGrid
```

```python
hori_frame = np.load('orthogonal pairs/hori.npy')
verti_frame = np.load('orthogonal pairs/verti.npy')
```

```python
fig =plt.figure(1, figsize=(60, 30))
grid=AxesGrid(fig, 236, nrows_ncols=(1, 2),
             axes_pad=0.5,
             share_all=False,
             cbar_location="right",
             cbar_mode="each",
             cbar_size="5%",
             cbar_pad="2%")
im=grid[0].imshow(hori_frame[0,:,:], cmap=plt.cm.viridis)
grid.cbar_axes[0].colorbar(im)
grid[0].set_title("horizontal frame")
im=grid[1].imshow(verti_frame[0,:,:], cmap=plt.cm.viridis)
grid.cbar_axes[1].colorbar(im)
grid[1].set_title("vertical frame")
```

```python
#plot the histogram
plt.rcParams['figure.figsize'] = (6.0, 3.0)
plt.rcParams['figure.dpi'] = 300
sns.histplot(data=(hori_frame.flatten(), verti_frame.flatten()), stat="density", kde=True)
```
